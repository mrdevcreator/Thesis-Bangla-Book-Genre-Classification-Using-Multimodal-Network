```python
import pandas as pd
import os

# Read the CSV file with image paths
df = pd.read_csv('Text+Image - Sheet1.csv')
df.dropna(inplace=True)

# Specify the folder containing all images
image_folder = 'All_image'

# Function to check if a file exists
def file_exists(filepath):
    return os.path.isfile(filepath)

# Lists to store results
missing_images = []
correct_images = []

# Counter for correct images
correct_images_count = 0

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    image_path = row['Image'].strip() if isinstance(row['Image'], str) else row['Image']  # Assuming the column with image paths is named 'Image'
    
    # Skip empty paths
    if pd.isna(image_path) or image_path == '':
        continue
    
    # Construct the full path
    full_path = os.path.join(image_folder, image_path)
    
    # Check if the file exists
    if file_exists(full_path):
        correct_images_count += 1
        correct_images.append(image_path)
    else:
        missing_images.append(image_path)

# Print results
print(f"Total images in CSV: {len(df)}")
print(f"Correct images found: {correct_images_count}")
print(f"Missing images: {len(missing_images)}")

# Print missing image paths
if missing_images:
    print("\nFirst 10 missing image paths:")
    for path in missing_images[:10]:
        print(path)
    if len(missing_images) > 10:
        print(f"... and {len(missing_images) - 10} more")

# Print correct image paths
print("\nFirst 10 correct image paths:")
for path in correct_images[:10]:
    print(path)
if len(correct_images) > 10:
    print(f"... and {len(correct_images) - 10} more")

# Find files in folder but not in CSV
files_in_folder = set(os.listdir(image_folder))
files_in_csv = set(df['Image'].dropna().apply(lambda x: x.strip() if isinstance(x, str) else x))
files_not_in_csv = files_in_folder - files_in_csv

print(f"\nFiles in folder but not in CSV: {len(files_not_in_csv)}")
print("First 10 files not in CSV:")
for file in list(files_not_in_csv)[:10]:
    print(file)
if len(files_not_in_csv) > 10:
    print(f"... and {len(files_not_in_csv) - 10} more")

# Optional: Save results to CSV files
pd.DataFrame({'MissingImagePath': missing_images}).to_csv('missing_image_paths.csv', index=False)
pd.DataFrame({'CorrectImagePath': correct_images}).to_csv('correct_image_paths.csv', index=False)
#pd.DataFrame({'FilesNotInCSV': list(files_not_in_csv)}).to_csv('files_not_in_csv.csv', index=False)

print("\nDetailed results saved to CSV files: 'missing_image_paths.csv', 'correct_image_paths.csv', and 'files_not_in_csv.csv'")
```
